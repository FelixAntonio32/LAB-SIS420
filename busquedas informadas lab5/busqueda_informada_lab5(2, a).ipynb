a. Conteo del numero de digitos que se encuentran en la posicion final deseada.

```python
def busqueda_heuristica(nodo_inicial, estado_objetivo, visitados, max_profundidad, profundidad=0):
    if profundidad > max_profundidad:
        return None  # Límite de profundidad alcanzado, terminar recursión

    visitados.append(nodo_inicial.get_estado())
    if nodo_inicial.get_estado() == estado_objetivo:
        return nodo_inicial
    else:
        estado_actual = nodo_inicial.get_estado()
        n = len(estado_actual)
        hijos = []

        for i in range(n - 1):
            hijo_estado = estado_actual[:i] + estado_actual[i + 1:i + 2] + estado_actual[i:i + 1] + estado_actual[i + 2:]
            hijos.append(NodoBI(hijo_estado, nodo_inicial))

        nodo_inicial.set_hijo(hijos)

        nodo_siguiente = None
        heuristica_max = -1

        for hijo_node in nodo_inicial.get_hijo():
            if hijo_node.get_estado() not in visitados:
                h = heuristica(nodo_inicial, hijo_node, estado_objetivo)
                if h > heuristica_max:
                    heuristica_max = h
                    nodo_siguiente = hijo_node

        if nodo_siguiente is not None:
            nodo_siguiente.set_padre(nodo_inicial)
            solu = busqueda_heuristica(nodo_siguiente, estado_objetivo, visitados, max_profundidad, profundidad + 1)
            if solu is not None:
                return solu
        return None

def heuristica(padre_node, hijo_node, estado_objetivo):
    padre_data = padre_node.get_estado()
    hijo_data = hijo_node.get_estado()
    objetivo = estado_objetivo

    heuristica_valor = sum(1 for i, j in zip(hijo_data, objetivo) if i == j)
    print(f"Conteo de dígitos en posición final deseada: {heuristica_valor}")

    return heuristica_valor
```

```python
def main():
    estado_inicial = [6, 5, 4, 3, 2, 1]
    estado_solucion = [1, 2, 3, 4, 5, 6]

    nodo_inicial = NodoBI(estado_inicial)
    max_profundidad = 100  # Establece aquí tu límite de profundidad deseado

    visitados_nodes = []
    nodo_solucion = busqueda_heuristica(nodo_inicial, estado_solucion, visitados_nodes, max_profundidad)

    if nodo_solucion:
        print("Se encontró una solución:")
        resultado = []

        node = nodo_solucion
        while node.get_padre() is not None:
            resultado.append(node.get_estado())
            node = node.get_padre()
        resultado.append(estado_inicial)
        resultado.reverse()

        conteo_digitos = heuristica(nodo_inicial, nodo_solucion, estado_solucion)
        print(f"Conteo de dígitos en posición final deseada: {conteo_digitos}")

        for estado in resultado:
            print(estado)
    else:
        print("No se encontró una solución.")

if __name__ == "__main__":
    main()
```
