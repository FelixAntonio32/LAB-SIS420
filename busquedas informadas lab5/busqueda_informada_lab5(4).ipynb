4. Proponer una nueva heuristica

```python
def heuristica_distancia_manhattan(estado_actual, estado_objetivo):
    heuristica_valor = 0

    for digit in estado_actual:
        if digit != 0:  # Excluye el espacio vacío si está presente
            actual_index = estado_actual.index(digit)
            objetivo_index = estado_objetivo.index(digit)
            distancia_filas = abs(actual_index // 3 - objetivo_index // 3)
            distancia_columnas = abs(actual_index % 3 - objetivo_index % 3)
            distancia_manhattan = distancia_filas + distancia_columnas
            heuristica_valor += distancia_manhattan

    return heuristica_valor


```

```python
## Define el estado objetivo
estado_objetivo = [1, 2, 3, 4, 5, 6, 7]

# Define el estado actual (ejemplo)
estado_actual = [7, 6, 5, 4, 3, 2, 1]

# Llama a la función heuristica_distancia_manhattan con los estados
heuristica_valor = heuristica_distancia_manhattan(estado_actual, estado_objetivo)

# Imprime el valor de la heurística
print(f"Valor de heurística de distancia de Manhattan: {heuristica_valor}")

# Imprime el estado actual
print("Estado actual:")
for i in range(0, len(estado_actual), 3):
    print(estado_actual[i:i + 3])

# Imprime el estado objetivo
print("Estado objetivo:")
for i in range(0, len(estado_objetivo), 3):
    print(estado_objetivo[i:i + 3])
```

la propuesta es la heuristica manhattan donde te permite calcular la distancia de la misma heuristica
