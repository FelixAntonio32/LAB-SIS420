1. Modificar el codigo fuente para resolver rompecabezas de N elementos.

```python
def busqueda_heuristica(nodo_inicial, estado_objetivo, visitados):
    visitados.append(nodo_inicial.get_estado())
    if nodo_inicial.get_estado() == estado_objetivo:
        return nodo_inicial
    else:
        # Expandir nodos sucesores (hijos)
        estado_actual = nodo_inicial.get_estado()
        n = len(estado_actual)
        hijos = []

        for i in range(n - 1):
            hijo_estado = estado_actual[:i] + estado_actual[i + 1:i + 2] + estado_actual[i:i + 1] + estado_actual[i + 2:]
            hijos.append(NodoBI(hijo_estado, nodo_inicial))

        nodo_inicial.set_hijo(hijos)

        for hijo_node in nodo_inicial.get_hijo():
            if hijo_node.get_estado() not in visitados and heuristica(nodo_inicial, hijo_node):
                hijo_node.set_padre(nodo_inicial)
                # Llamada recursiva
                solu = busqueda_heuristica(hijo_node, estado_objetivo, visitados)
                if solu is not None:
                    return solu
        return None

def heuristica(padre_node, hijo_node):
    padre_quality = 0
    hijo_quality = 0
    padre_data = padre_node.get_estado()
    hijo_data = hijo_node.get_estado()

    for n in range(1, len(padre_data)):
        if padre_data[n] > padre_data[n - 1]:
            padre_quality += 1
        if hijo_data[n] > hijo_data[n - 1]:
            hijo_quality += 1

    return hijo_quality >= padre_quality
```

```python
def main():
    estado_inicial = [7, 6, 5, 4, 3, 2, 1]
    estado_solucion = [1, 2, 3, 4, 5, 6, 7]

    visitados_nodes = []
    nodo_inicial = NodoBI(estado_inicial)
    nodo_solucion = busqueda_heuristica(nodo_inicial, estado_solucion, visitados_nodes)

    if nodo_solucion:
        print("Se encontró una solución:")
        resultado = []

        node = nodo_solucion
        while node.get_padre() is not None:
            resultado.append(node.get_estado())
            node = node.get_padre()
        resultado.append(estado_inicial)
        resultado.reverse()

        for estado in resultado:
            print(estado)
    else:
        print("No se encontró una solución.")

if __name__ == "__main__":
    main()
```
